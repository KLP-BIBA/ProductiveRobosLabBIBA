```python
 
import ipywidgets as widgets
from IPython.display import display
import append_text

# Text input for file name
file_input = widgets.Text(
    value='../meshes/Fences/FencesGenerated.urdf',
    placeholder='Enter file name',
    description='File:',
)

# Multiline text input for snippet
text_input = widgets.Textarea(
    value='Hello from Jupyter!',
    placeholder='Enter text to append',
    description='Text:',
    layout=widgets.Layout(width='500px', height='150px')  # optional sizing
) 

# Button to trigger action
button = widgets.Button(description="Append Text")

# Output area
output = widgets.Output()

def on_button_click(b):
    with output:
        output.clear_output()
        file_path = file_input.value
        text_snippet = text_input.value
        
        append_text.append_to_file(file_path, text_snippet)
        print(f"   ")
        print(f"Appended to: {file_path}")
        
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
        
        print(content)

button.on_click(on_button_click)

display(file_input, text_input, button, output)
```
