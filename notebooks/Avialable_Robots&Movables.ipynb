```python
from IPython.display import display, Markdown, IFrame
from ipywidgets import Button, Layout, GridBox
from pathlib import Path
import rospy
from sidecar import Sidecar
from helper import Launcher, Robots, Rvizweb

RVIZWEB_URL = url = rospy.get_param('/rvizweb/jupyter_proxy_url')
IFRAME = IFrame(src=RVIZWEB_URL, width='100%', height='100%')
def MAKE_SIDECAR():
    return Sidecar(title='RVIZWEB', anchor='right')
RVIZWEB = Rvizweb(make_sidecar=MAKE_SIDECAR, iframe=IFRAME, display=display)
ROBOTS = Robots()
LAUNCHER = Launcher(Path.home() / 'work/launch/robots.launch')
```

```python
ROBOTS.register('UR5 6axis', Path.home() / 'work/meshes/UR5/ur5.urdf')
ROBOTS.register('UR3 6axis', Path.home() / 'work/meshes/UR3/ur3.urdf')
ROBOTS.register('UR10 6axis', Path.home() / 'work/meshes/UR10/ur10.urdf')
ROBOTS.register('CR35ia 6axis', Path.home() / 'work/meshes/CR35ia/cr35ia.urdf')
ROBOTS.register('r2000ib 210f 6axis', Path.home() / 'work/meshes/R2000ib210f/r2000ib210f.urdf')
ROBOTS.register('RG2 Gripper', Path.home() / 'work/meshes/Grippers/OnrobotRG2/OnrobotRG2.urdf')
ROBOTS.register('SoftGripper', Path.home() / 'work/meshes/Grippers/OnrobotSoftgripper/OnrobotSG_b.urdf')
ROBOTS.register('Dual Gripper', Path.home() / 'work/meshes/Grippers/OnrobotDualQuickChanger/OnrobotDualChangeWithRG2.urdf', '<b>Dual QuickChanger</b><br>Onrobot Dual-Quickchanger is visualized with Gripper and Softroobt')
```

### Code to run

```python

def create_button(name):
    btn = Button(
        description=name,
        layout=Layout(width='auto', height='50px'),
        style={'font_size':'1rem'},
        tooltip=f"Launch robot: {name}"
    )
    def launch_robot(name):
        urdf_file = ROBOTS.get_urdf_file(name)
        if urdf_file:
            print(f'Launch robot {name} with urdf file {str(urdf_file)}')
            LAUNCHER.launch(urdf_file=urdf_file,
                            process_name=name,
                            rvizweb=RVIZWEB)
        else:
            print(f'No urdf file for robot "{name}" available')
    btn.on_click(lambda b: launch_robot(name))
    return btn

if LAUNCHER.get_process_name() != None:
    display(Markdown(f'### Robot **{LAUNCHER.get_process_name()}** Launched'))
display(Markdown('#### Click the buttons below to launch an industrial robot.'))

display(ROBOTS.get_selection_widget(create_button))
```
